```python
from ultralytics import YOLO
import torch

# Inicialización de modelo pre-entrenado base
model = YOLO("yolov8n.pt")

# Ejecución de entrenamiento con hiperparámetros optimizados para LSCh
model.train(
    data="dataset/data.yaml",   # Ubicación del conjunto de datos en formato YOLO
    epochs=100,         # Épocas incrementadas para optimizar convergencia en clasificador de 26 clases
    imgsz=640,          # Resolución estándar de fotograma
    batch=20,           # Tamaño de lote incrementado dada la ausencia de superposición compleja
    lr0=0.001,          # Tasa de aprendizaje reducida para favorecer convergencia estable
    lrf=0.01,           # Tasa de aprendizaje final (decaimiento)
    momentum=0.937,     # Momento paramétrico para el optimizador
    weight_decay=0.0005, # Factor de regularización para mitigar sobreajuste
    warmup_epochs=3,    # Calentamiento para estabilizar la fase inicial de entrenamiento
    warmup_momentum=0.8,
    optimizer="AdamW",  # Optimizador AdamW (preferente para conjuntos de datos de escala moderada)
    cos_lr=True,        # Planificador de tasa de aprendizaje cosenoidal
    device=0,           # Asignación a unidad de procesamiento gráfico (GPU)
    patience=15,        # Criterio de detención temprana ante ausencia de mejoras
    save_period=10,     # Frecuencia de respaldo de puntos de control (checkpoints)
    val=True,           # Ejecución de validación intra-entrenamiento
    plots=True,         # Habilitación de generación de métricas gráficas
    verbose=True,       # Habilitación de salida de registro detallada (verbose)
    seed=42,            # Semilla de aleatoriedad para garantizar reproducibilidad
    deterministic=True, # Configuración determinista de operaciones
    workers=4,          # Nivel de concurrencia para carga de datos
    rect=True,          # Entrenamiento rectangular para optimización de eficiencia
    # Parámetros de superposición excluidos intencionadamente
    dropout=0.0,        # Sin dropout inicial
    label_smoothing=0.1, # Suavizado de labels para mejor generalización
    # Parámetros específicos para detección de una mano
    single_cls=False,   # Múltiples clases (A-Z)
    save_txt=True,      # Guardar predicciones en texto
    save_conf=True,     # Guardar confianzas
    # Augmentaciones apropiadas para manos
    hsv_h=0.015,        # Variación de tono (lighting)
    hsv_s=0.7,          # Variación de saturación
    hsv_v=0.4,          # Variación de brillo
    degrees=10,         # Rotación máxima (manos pueden rotar ligeramente)
    translate=0.1,      # Traslación (mano puede moverse en frame)
    scale=0.2,          # Escala (mano más cerca/lejos)
    shear=2,            # Menor shear para mantener forma de mano
    perspective=0.0002, # Mínima perspectiva
    flipud=0.0,         # Sin flip vertical (manos tienen orientación)
    fliplr=0.5,         # Flip horizontal OK (mano izq/der)
    mosaic=0.8,         # Mosaic augmentation para variedad
    mixup=0.1,          # Mixup ligero para generalización
)
```

```python
model = YOLO("runs/detect/train/weights/best.pt")

# 2. Usar la cámara (source=0 es la webcam principal)
model.predict(
    source=0,        # webcam
    show=True,       # mostrar ventana en tiempo real
    conf=0.5,        # confianza mínima (ajústala si detecta de más o de menos)
    save=False       # si quieres guardar el video, pon True
)
```
